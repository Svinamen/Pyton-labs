- Варіант 6:

$$-10.3x_{1}+10.2x_{2} = 70\\4.7x_{1}+12.3x_{2}=173\\13.2x_{1}+8.8x_{2} = 282$$


```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-bright')
```

```python
fig = plt.figure(0,(9.6,12))
x = np.linspace(0,50,25)
x_1=(-70/10.3)+(10.2/10.3)*x
x_2=(173/4.7)+(12.3/4.7)*x
x_3=(282/13.2)+(8.8/13.2)*x
y_n = 50
x_n = 20

fig.add_subplot(3,2,1)
plt.title("x_1=(-70/10.3)+(10.2/10.3)*x")
plt.plot(x,x_1, color = "#00ff00", linestyle='--')
plt.axis([5,x_n,0,y_n])

fig.add_subplot(3,2,2)
plt.title("x_2=(173/4.7)+(12.3/4.7)*x")
plt.plot(x,x_2, 'y:')
plt.axis([0,5,0,y_n])


fig.add_subplot(3,2,3)
plt.title("x_3=(282/13.2)+(8.8/13.2)*x")
plt.plot(x,x_3, color = (0.5,0.2,1), linestyle = '-.')
plt.axis([0,x_n,0,y_n])

plt.show()
```

```python
fig1 = plt.figure(0,(10,8),200)
x = np.linspace(0,20,25)

plt.plot(x,(70+10.3*x)/10.2, color = "#008800", linestyle='--', label = "x_1= -10.3x1+10.2x2 = 70")
plt.plot(x,(173-4.7*x)/12.3, 'b:', label = "x_2=4.7x1+12.3x2=173")
plt.plot(x,(282-13.2*x)/8.8, color = (0.9,0.1,0.6), linestyle = '-.', label = "x_3=13.2x1+8.8x2 = 282")
plt.text(12, 20, 'x_1= -10.3x1+10.2x2 = 70', rotation = 27, fontsize = 12)
plt.text(0, 12.7, 'x_2=4.7x1+12.3x2=173', rotation = -10, fontsize = 12)
plt.text(2.5, 21.7, 'x_3=13.2x1+8.8x2 = 282', rotation = -38, fontsize = 12)
plt.legend()
a = np.minimum((70+10.3*x)/10.2,(282-13.2*x)/8.8)
plt.fill_between(x,(173-4.7*x)/12.3,a, where = (x<16.4) & (x>4.25), alpha=0.5, color = "#ff9966")
plt.title("Function intersection graph")
plt.ylabel("0 < y < 30")
plt.xlabel("0 < x < 20")
plt.grid()
print(np.linalg.solve([[-10.3,10.2],[4.7,12.3]],[70,173]))
print(np.linalg.solve([[-10.3,10.2],[13.2,8.8]],[70,282]))
print(np.linalg.solve([[4.7,12.3],[13.2,8.8]],[173,282]))
plt.plot(16.08,7.91,'ro')
plt.plot(5.17,12.08, 'ro')
plt.plot(10.03,17, 'ro')

fig1.savefig('function_intersection_graph.png', dpi = 1000)
fig1.savefig('function_intersection_graph.jpg', dpi = 1000)
fig1.savefig('function_intersection_graph.svg', dpi = 1000)

plt.show()
```
